```python
import requests
import pandas as pd
import io

url = 'https://raw.githubusercontent.com/kaharmc226/classification-data/refs/heads/main/Stress%20Dataset.csv'
response = requests.get(url)
response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
data = io.StringIO(response.text)

df = pd.read_csv(data)
```
